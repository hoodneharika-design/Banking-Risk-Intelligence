```python
import pandas as pd

# Load original dataset
df = pd.read_csv(r'data\creditcard.csv')

# Take 5000 rows only
small_df = df.sample(5000, random_state=42)

# Save smaller dataset
small_df.to_csv(r'data\creditcard_sample.csv', index=False)

print("✅ Smaller dataset saved!")
print("Shape:", small_df.shape)
```
